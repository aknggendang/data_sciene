# Data Type Convert


```python
anime_ds["episodes"]
```

```python
anime_ds["episodes"].tail(40)
```

Replace 'Unknown' Values to Zero values in Episodes Column 

```python
anime_ds['episodes'] = anime_ds['episodes'].replace('Unknown',0)
```

Convert Episodes Column from object data type to int data type

```python
anime_ds['episodes'] = pd.to_numeric(anime_ds['episodes']).astype(int)
```
